```python
# импортируем библиотеку
import pandas as pd
# применим функцию read_csv() и посмотрим на первые три записи файла Trending videos on youtube dataset.csv
Trend_vid = pd.read_csv('/content/Trending videos on youtube dataset.csv')
Trend_vid.head(10)
```

```python
Trend_vid.info()
```

```python
# для построения графиков воспользуемся новой для нас библиотекой seaborn
import seaborn as sns
# посмотрим влияет ли качество видео на количество просмотров
# с помощью x и hue мы можем уместить две категориальные переменные на одном графике
sns.countplot(x = 'viewCount', hue = 'definition', data = Trend_vid.head(10))
```

```python
# для построения графиков воспользуемся новой для нас библиотекой seaborn
import seaborn as sns
# посмотрим влияет ли качество видео на количество дизлайков
# с помощью x и hue мы можем уместить две категориальные переменные на одном графике
sns.countplot(x = 'dislikeCount', hue = 'definition', data = Trend_vid.head(10))
```

```python
Trend_vid.isnull().sum()
```

```python
# переменная videoDescription (описание видео), скорее всего, не является самой важной
# избавимся от нее с помощью метода .drop()
# (параметр axis отвечает за столбцы, inplace = True сохраняет изменения)
Trend_vid.drop(columns = 'videoDescription', axis = 1, inplace = True)
```

```python
# а вот dislikeCount (количество дизлайков) точно важен, заменим пустые значения средним арифметическим
Trend_vid['dislikeCount'].fillna(Trend_vid['dislikeCount'].mean(), inplace = True)
```

```python
# у нас остаются пустые строки в commentCount   likeCount, удалим их
Trend_vid.dropna(inplace = True)
```

```python
# применим one-hot encoding к переменной definition (четкость) с помощью метода .get_dummies()
pd.get_dummies(Trend_vid['definition']).head(3)
```

```python
# удалим первый столбец, он избыточен
definition = pd.get_dummies(Trend_vid['definition'], drop_first = True)
definition.head(3)
```

```python
# применим one-hot encoding к переменной caption (Подпись) с помощью метода .get_dummies()
caption = pd.get_dummies(Trend_vid['caption'], drop_first = True)
caption.head(3)
```

```python
Trend_vid = pd.concat([Trend_vid, caption,definition], axis = 1)
```

```python
# применим функцию drop() к соответствующим столбцам
Trend_vid.drop(['channelId', 'videoTitle', 'videoId', 'publishedAt',
'videoCategoryId', 'videoCategoryId','videoCategoryLabel','duration','definition','channelTitle','caption'], axis = 1, inplace = True)
Trend_vid.head(3)
```

```python
# импортируем класс StandardScaler
from sklearn.preprocessing import StandardScaler
# создадим объект этого класса
scaler = StandardScaler()
# выберем те столбцы, которые мы хотим масштабировать
cols_to_scale = ['durationSec', 'viewCount','likeCount','commentCount','dislikeCount']
# рассчитаем среднее арифметическое и СКО для масштабирования данных
scaler.fit(Trend_vid[cols_to_scale])
# применим их
Trend_vid[cols_to_scale] = scaler.transform(Trend_vid[cols_to_scale])
# посмотрим на результат
Trend_vid.head(3)
```

```python
Trend_vid.columns = Trend_vid.columns.map(str)
```

```python
# поместим в X_train все кроме столбца sd
X_Trend_vid = Trend_vid.drop('sd', axis = 1)
# столбец 'sd' станет нашей целевой переменной (y_train)
y_Trend_vid = Trend_vid['sd']
X_Trend_vid.head(3)
```

```python
# импортируем логистическую регрессию из модуля linear_model библиотеки sklearn
from sklearn.linear_model import LogisticRegression
# создадим объект этого класса и запишем его в переменную model
model = LogisticRegression()
# обучим нашу модель
model.fit(X_Trend_vid, y_Trend_vid)
LogisticRegression()
```

```python
# сделаем предсказание класса на обучающей выборке
y_pred_Trend_vid = model.predict(X_Trend_vid)
```

```python
# построим матрицу ошибок
from sklearn.metrics import confusion_matrix
# передадим ей фактические и прогнозные значения
conf_matrix = confusion_matrix(y_Trend_vid, y_pred_Trend_vid)
# преобразуем в датафрейм
conf_matrix_df = pd.DataFrame(conf_matrix)
conf_matrix_df
```

```python
conf_matrix_labels = pd.DataFrame(conf_matrix, columns =
['Прогноз hd', 'Прогноз sd'], index = ['Факт hd', 'Факт sd'])
conf_matrix_labels
```
